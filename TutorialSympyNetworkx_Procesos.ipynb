#$\color{darkred}{\text{TUTORIAL SYMPY PARA MULTIPLICAR MATRICES}}$

Se importa la libreria sympy para poder hacer multiplicacion de matrices

```python
from sympy import *
init_printing(use_unicode=True)
```

Para crear una matriz en SymPy, se utiliza $$\textbf{Matrix}([ \quad ])$$

Una matriz se construye proporcionando una lista de vectores fila que la componen.

Dentro de los corchetes irГЎn las filas de cada matriz igual entre corchetes separando cada entrada por una coma, por ejemplo: $$[1 , -1]$$ esa es la primer fila.

Para introducir una segunda fila, despuГ©s del corchete de la fila anterior se coloca una coma y se abren corchetes nuevamente para ingresar la segunda fila, por ejemplo: $$[1,-1] , [3, 4]$$

AquГ­ ya se tienen dos filas de la matriz.

Finalmente la forma completa serГ­a

$$\textbf{Matrix}([ [1 , -1] , [3, 4] ])$$

Lo anterior imprime una matriz $2 $ x $ 2$, pero se puede poner una matriz de dimensiГіn $n$ x $n$ con $n$ un nГєmero entero.

Por ejemplo:

```python
Matrix([[1, -1, 3], [3, 4, 5], [0, 2, 7]])
```

Para llamar a la Matriz anterior se le puede asignar una letra, con esto la nombramos y ya no es necesario escribir toda la matriz en casos posteriores, basta con Гєnicamente llamarla por el nombre que se le puso.

Por ejemplo:

```python
M=Matrix([[1, -1, 3], [3, 4, 5], [0, 2, 7]])
```

Le asigne el nombre M, pero pudo haber sido cualquier otro, ahora basta con escribir su nombre en la lГ­nea de cГіdigo para llamarla:

```python
M
```

Ahora para multiplicar matrices, basta con usar el operador "*" entre las matrices que se desea multiplicar, por ejemplo:

$$M \text{*} N$$

Con M una matriz de $n \text{x} m$ dimensiones y N una matriz de $m \text{x} p$ dimensiones. Las dimensiones son importantes para saber si es posible multiplicar matrices o vectores.

Ejemplo:

```python
N=Matrix([[1,  3], [3,  5], [0, 7]])
```

Se crea una matriz N de $3 \text{x} 2$ dimensiones, que es compatible con la matriz M que se creo anteriormente

```python
N
```

Ahora se puede llevar acabo la multiplicaciГіn

```python
M*N
```

Se tiene el resultado de la multiplicaciГіn, que nuevamente es una matriz

De igual manera se puede obtener una fila, columna o entrada especifica de una matriz, esto de la forma

$$M[i,:]$$

Se obtiene la fila $i.$

```python
M[1,:]
```

Para la columna es de manera similar:

$$M[:,i]$$

Se obtiene la columna $i$.

```python
M[:,0]
```

Y para obtener una entrada especifica, se pone la posiciГіn en la que se encuentra:

$$M[i,j]$$

Se obtiene la entrada $i,j \text{ con }\quad i,j=0,1,2,...n-1$

```python
M[2,2]
```

Elevar una matriz a una potencia dada es de la forma:

$$M\text{**}n \text{ con } n=0,1,2,...$$

```python
M**2
```

Ahora para obtener valores propios de una matriz, es con el comando

eigenvals()

de la forma M.eigenvals() donde M es la matriz

```python
M.eigenvals()
```

Para obtener los vectores propios se usa eigenvects()

De una manera similar M.eigenvects() donde M es el nombre de la Matriz

```python
M.eigenvects()
```

El resultado depende de cada Matriz, en este caso fue un resultado "complejo", pero eso varea de acuerdo a cada Matriz

# $\color{darkred}{\text{TUTORIAL DE NETWORKX PARA DIBUJAR DIAGRAMAS DE TRANSICIГ“N}}$

Se importa la libreria networkx

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

Para crear un grГЎfico vacio es:

```python
G = nx.Graph()
```

El grafo G se puede hacer crecer de varias maneras. NetworkX incluye muchas funciones de generador de grafos y facilidades para leer y escribir grafos en muchos formatos. Sin embargo, para empezar, veremos manipulaciones sencillas. Puedes aГ±adir un nodo a la vez,

```python
G.add_node(1)
```

o agregar nodos desde cualquier contenedor iterable , como una lista

```python
G.add_nodes_from([2, 3])
```

TambiГ©n puede agregar nodos junto con atributos de nodo si su contenedor produce 2 tuplas del formato

```python
G.add_nodes_from([(4, {"color": "red"}), (5, {"color": "green"})])
```

Los nodos de un grГЎfico se pueden incorporar a otro:

```python
H = nx.path_graph(10)
G.add_nodes_from(H)
```

G Ahora contiene los nodos de H como nodos de G. Por el contrario, podrГ­as usar el grГЎfico H como un nodo en G.

```python
G.add_node(H)
```

El grГЎfico Gahora contiene Hun nodo. Esta flexibilidad es muy potente, ya que permite crear grГЎficos de grГЎficos, grГЎficos de archivos, grГЎficos de funciones y mucho mГЎs.

#Ejemplo Sencillo

Para crear un grГЎfico dirigido es con:

DiGraph()

```python
# Crear grafo dirigido
G = nx.DiGraph()
```

Supongamos que tenemos 3 estados: A, B, C, y queremos indicar probabilidades de transiciГіn:

```python
# Agregar transiciones
G.add_edge("A", "B", weight=0.6)
G.add_edge("B", "C", weight=0.4)
G.add_edge("C", "A", weight=1.0)
```

Podemos dibujar usando nx.draw y mostrar las etiquetas de las aristas (transiciones):

```python
# PosiciГіn automГЎtica de los nodos
pos = nx.spring_layout(G)

# Dibujar grafo y etiquetas de las aristas
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G,'weight'))

plt.show()
```

Se puede mejorar el diagrama poniendo titulo, color a los nodos, aumentar el tamaГ±o, de la misma manera en que se grГЎfica comunmente.

```python
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500, arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G,'weight'))
plt.title("Diagrama de TransiciГіn")
plt.show()
```

Un ejemplo muy simple.

Pero ahora, dada una matriz de transiciГіn se crearГЎ el diagrama

Se crea un arreglo numpy para facilitar el grafo

```python
P = np.array([[0, 0.6, 0.4],
              [0, 0.5, 0.5],
              [0.8, 0.2, 0]])
```

Se les da nombres a los estados

```python
# Nombres de los estados
estados = ["A", "B", "C"]
```

Se usa un comando de networkx y numpy para crear el diagrama a partir de la matriz

```python
G = nx.from_numpy_array(P, create_using=nx.DiGraph)  # crea el grafo desde la matriz
```

Finalmente se grГЎfica de la misma manera

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.show()
```

O bien, se pueden agregar de manera mГЎs manual incluyendo todo lo deseado, de la manera siguiente:

```python
# Crear grafo dirigido
G = nx.DiGraph()

# Agregar aristas con probabilidades
for i, fila in enumerate(P):
    for j, prob in enumerate(fila):
        if prob > 0:
            G.add_edge(estados[i], estados[j], probabilidades=prob)
```

Se grГЎfica de la misma manera, incluyendo ahora las probabilidades

```python
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True)

# Dibujar etiquetas de las aristas con las probabilidades
edge_labels = nx.get_edge_attributes(G, 'probabilidades')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.show()
```
